```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier ,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y=make_classification(n_features=5,n_redundant=0,n_informative=1,n_clusters_per_class=1)
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
print(df.shape)
df.head()
```

```python
bag=BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,0:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,9))
plot_tree(bag.estimators_[0])
plt.show
```

```python
ran=RandomForestClassifier(max_features=2)
ran.fit(df.iloc[:,0:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(14,9))
plot_tree(ran.estimators_[4])
plt.show
```
