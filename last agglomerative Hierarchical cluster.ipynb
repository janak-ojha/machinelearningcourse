```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cd=pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
cd
```

```python
newcd=cd.iloc[:,3:5].values
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(14,7))
plt.title('customer_dendogram')
dend=shc.dendrogram(shc.linkage(newcd,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
labels_=cluster.fit_predict(newcd)
```

```python
labels_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(newcd[:,0],newcd[:,1],c=cluster.labels_,cmap='rainbow')
```
